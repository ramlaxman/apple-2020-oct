# Exercise: Books as named tuples

1. Using `namedtuple`, create a `Book` class with attributes `title`, `author`, and `price`.
2. Using this new `Book` class, define three different books, and put them in a list of books the store sells.
3. Allow a user/customer to enter the name of a book they want to buy.
    - If the user gives us an empty string, then we stop asking.
    - If they name a book that we do carry, then we'll print its details and add the price to the total
    - If they name a book that we do *not* carry, then we'll tell them so
4. After they're done asking for books, we'll print the total price.

```python
from collections import namedtuple

Book = namedtuple('Book', ['title', 'author', 'price'])

all_books = [Book('a', 'author 1', 50),
            Book('b', 'author 2', 30),
            Book('c', 'author 2', 60)]

total = 0

while True:
    title = input("Enter a book title: ").strip()
    
    if not title:
        break
        
    for one_book in all_books:
        if one_book.title == title:
            print(f'Found {one_book}')
            total += one_book.price
            print(f'\tTotal is now {total}')
            break

    else:    # got to the end of the "for" loop without any "break"
        print(f'Did not find {title}')
            
print(f'Total is {total}')
```

```python
from collections import namedtuple

Book = namedtuple('Book', ['title', 'author', 'price'])

all_books = [Book('a', 'author 1', 50),
            Book('b', 'author 2', 30),
            Book('c', 'author 2', 60)]

total = 0

# := is the "assignment expression" operator
#  "walrus" -- from 3.8 and onwards

while title := input("Enter a book title: ").strip():
    
    for one_book in all_books:
        if one_book.title == title:
            print(f'Found {one_book}')
            total += one_book.price
            print(f'\tTotal is now {total}')
            break

    else:    # got to the end of the "for" loop without any "break"
        print(f'Did not find {title}')
            
print(f'Total is {total}')
```
